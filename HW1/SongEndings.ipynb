```python
#create searchable dictionaries of endings from an input file
def load_endings(filename):
    #open endings file
    endings_file = open(filename, "r")

    #create blank dictionary to search by ending
    endings = {}
    
    #create blank dictionary to search by end_sound
    end_sounds = {}

    #iterate through file
    for line in endings_file:
        #clean line and split by ::
        data = str(line).rstrip().split("::")

        #seperate sound and other info
        end_sound = data[1]
        ending_info = [data[0], int(data[2])]
        ending = data[0]
        ending_data = [data[1], int(data[2])]
        
        #populate endings dictionary
        endings[ending] = ending_data
        
        #populate end_sounds dictionary
        #check if sound is already in dictionary
        if end_sound in end_sounds:
            #append ending info to sound key
            end_sounds[end_sound].append(ending_info)
            
        else:
            #create sound as key and append ending info
            end_sounds[end_sound] = []
            end_sounds[end_sound].append(ending_info)

    #close endings file
    endings_file.close
    
    #return loaded endings dictionary
    return endings, end_sounds

#complete a song with rhyming endings from a given dictionary
def complete_song(filename, endings, end_sounds):
    #open song file
    song_file = open(filename, "r")
    
    #create a blank string to store the final song
    song = """"""
    
    #iterate through song file
    for line in song_file:
        
        #clean and split string
        data = str(line).rstrip().split("::")
        
        #separate string parts
        prefix = data[0]
        prefix_beats = int(data[1])
        ending = data[2]
        
        #check if ending is blank
        if ending == "":
            #add line to song
            song += prefix + "\n"
        
        #check if ending is incomplete
        elif ending == "XXX":
            #get needed beats
            beats = previous_beats - prefix_beats
            
            #get list of possible rhymes
            possible_rhymes = end_sounds[previous_end_sound]
            
            #loop through rhymes
            for ending_info in possible_rhymes:
                
                #save the last ending with the appropriate number of beats
                if beats == ending_info[1]:
                    new_ending = ending_info[0]
            #add line to song
            song += prefix + " " + new_ending + "\n"
            
        #if ending is not blank or incomplete, create "previous line" data
        else:
            #save line beats
            previous_beats = prefix_beats + endings[ending][1]
            
            #save line ending
            previous_end_sound = endings[ending][0]
            
            #add line to song
            song += prefix + " " + ending + "\n"
    
    #close song file
    song_file.close()
    
    #return song
    return song

#run the load_endings function
endings, end_sounds = load_endings("endings.txt")

#run the complete_song function
song = complete_song("skeleton_SamuraiShowdown.txt", endings, end_sounds)

print song
            
```
