DOING DECISION TREE CLASSIFICATION!!!



```python
from patsy import dmatrices
from sklearn import tree
import warnings
%pylab inline
warnings.filterwarnings('ignore')
```

```python
df['target'] = 0.0
mask = (df['time_in_shelter_days'] < 16.7571155453)
df['target'][mask] = 1.0
df
```

```python
formula = 'target ~ 0 + outcome_type + age_upon_intake + animal_type + breed + color + ' + \
          'intake_condition + intake_type + intake_month'
    
Y, X = dmatrices(formula, df, return_type='dataframe')
y = Y['target'].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
```

```python
from sklearn.model_selection import cross_val_score, StratifiedKFold

# Create the folds in the training data
kfold = StratifiedKFold(n_splits=3, shuffle=True)

# Iterate over max_depth
for max_depth in [1, 2, 3, 4, 5]:
    model3 = tree.DecisionTreeClassifier(criterion='entropy', max_depth=max_depth)
    scores = cross_val_score(model3, X_train, y_train, cv=kfold)
    print 'max_depth={} scores={} avg_score={}'.format(max_depth, scores, scores.mean())
    
    
```

Chosing to so max_depth of 7 (after this is seems to increase by 0.02 each time and gets too big) 

```python
model7 = tree.DecisionTreeClassifier(criterion='entropy', max_depth=4)
result = model7.fit(X_train, y_train)
```

```python
from StringIO import StringIO
import pydot_ng
dot_data = StringIO()
tree.export_graphviz(model7, out_file=dot_data, feature_names=X.columns.values)
pydot_ng.graph_from_dot_data(dot_data.getvalue()).write_png('Average_Time_In_Shelter.png')
```

```python
def determine_season(item):
    if item in [12,1,2]:
        return "Winter"
    elif item in [3,4,5]:
        return "Spring"
    elif item in [6,7,8]:
        return "Summer"
    else:
        return "Fall"
    
    
df['intake_season'] = df['intake_month'].map(determine_season)
df['outcome_season'] = df['outcome_month'].map(determine_season)
df
    
    
```
