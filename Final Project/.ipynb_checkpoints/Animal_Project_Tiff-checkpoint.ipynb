```python
#put age into buckets
import re

def get_age_bucket(val):
    if re.search("day", val):
        return "Less than 1 week"
    elif re.search("week", val):
        return "1-6 weeks"
    elif re.search("[1,2,3,4,5,6] month", val):
        return "1-6 months"
    elif re.search("[7,8,9,10,11,12] month", val):
        return "7-12 months"
    elif re.search("[1,2,3] year", val):
        return "1-3 years"
    elif re.search("[4,5,6] year", val):
        return "4-6 years"
    elif re.search("year", val):
        return "7+ years"
    else:
        return "other"

df['Age_Bucket'] = df['age_upon_outcome'].apply(get_age_bucket)
df['Age_Bucket'].value_counts()
```

```python
#how long were animals at shelter
#put time_in_shelter into buckets

import re

def get_days_length(val):
    val = str(val)
    days = re.findall('\d*',val)[0]
    try:
        days = int(days)
        if days <= 7:
            return "0-7 days"
        elif days <= 21:
            return "1-3 weeks"
        elif days <= 42:
            return "3-6 weeks"
        elif days <= 84:
            return "7-12 weeks"
        elif days <= 168:
            return "12 weeks - 6 months"
        elif days <= 365:
            return "6-12 months"
        elif days <= 730:
            return "1-2 years"
        else:
            return "2+ years"
    except:
        return np.nan

df['Days_length'] = df['time_in_shelter_days'].apply(get_days_length)
vc_time = df['Days_length'].value_counts()
vc_time
```

```python
#what is the likelihood of outcome based on age?


dogs_outcome_age = dogs.groupby(['outcome_type','Age_Bucket'])['Age_Bucket'].count().unstack()
# Drop outcomes we're not interested in
dogs_outcome_age = dogs_outcome_age.drop(dogs_outcome_age.index[[1,2,4,6]])

# Transform into a percentage
def get_perc_outcome(val,col):
    return float(val) / float(dogs_outcome_age[col].sum()) * 100

perc_dog_outcome_age = pd.DataFrame()
for col in dogs_outcome_age.columns:
    perc_dog_outcome_age[col] = dogs_outcome_age[col].apply(get_perc_outcome, args=(col,))

perc_dog_outcome_age = perc_dog_outcome_age.T
perc_dog_outcome_age.plot(kind="bar", figsize=(20,5), title = "Distribution of outcome types based on age group")
```

```python
#can we predict adoption vs. other outcomes?

df['Target'] = 0
adoption_mask = df['outcome_type'] == 'Adoption'
df.loc[adoption_mask, 'Target'] = 1
df[:20]
```

```python
#what is our baseline prediction accuracy?

majority = float(df['Target'].value_counts()[0])
total = df['Target'].value_counts().sum()
baseline = majority / total
print baseline

df['Target'].value_counts()
```

```python
#let's try nearest neighbors

from patsy import dmatrices
from sklearn.model_selection import train_test_split
from sklearn import neighbors
from sklearn import metrics

y, X = dmatrices('Target ~ 0 + Gender + age_upon_intake + animal_type + breed + intake_condition + Days_length', df, return_type = 'dataframe')
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
```

```python
neighbor_list = [20,50,200]

for neighbor_num in neighbor_list:
    model = neighbors.KNeighborsClassifier(n_neighbors=neighbor_num,
                                       weights='uniform',
                                       p=2)
    
    print 'Number of Neighbors:', neighbor_num
    model.fit(X_train, y_train.values.ravel())
    prediction_on_training = model.predict(X_train)
    print 'Train accuracy:', metrics.accuracy_score(y_train, prediction_on_training)
    prediction_on_test = model.predict(X_test)
    print 'Test accuracy:', metrics.accuracy_score(y_test, prediction_on_test)
    print ''
```
