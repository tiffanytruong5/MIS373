```python
#Rock, Paper, Scissors Game
#Created a matrix with all potential combinations of computer and user choices, as well as a series to keep track of the player's decisions
#Used a while loop to keep the code running as the user plays more games

from pandas import Series, DataFrame
import random
import pandas as pd

#Set up dataframe with all possible combinations, column is computer choice and row is human choice
options = ['Rock','Paper','Scissors']
data = {'Rock':['Tied','Lost','Won'],'Paper':['Won','Tied','Lost'],'Scissors':['Lost','Won','Tied']}
df = DataFrame(data, index = options)

#Set up series keeping track of player's decisions.
count = [0,0,0]
history = Series(count, index = options)

wins = 0
games = 0

#Conditions to whether or not player wants to keep playing
playing = True
while playing == True:
    player_choice = raw_input('Your move: ')
    if (player_choice not in options) and (player_choice != "Quit"):
        print('Invalid decision, try again: ')
    elif (player_choice == "Quit"):
        playing = False
        print ("You played "+str(games)+" and won "+str(wins))
    else:
        if history.idxmax() == "Rock":
            comp_choice = "Paper"
        elif history.idxmax() == "Paper":
            comp_choice = "Scissors"
        else:
            comp_choice = "Rock"
        result = df.get_value(comp_choice,player_choice)
        print ("You chose "+player_choice+", the computer chose "+comp_choice+": You "+result)
        if result == "Won":
            wins +=1
        games +=1
        history[player_choice] +=1
```
