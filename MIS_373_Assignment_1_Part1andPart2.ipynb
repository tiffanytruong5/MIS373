```python
import random

#define global variables
ROCK = 1
PAPER = 2
SCISSORS = 3
QUIT = 4


HUMAN = 1
COMPUTER = 0
TIE = - 1


def main():


    #set variables
    num_games = 0
    num_human_wins = 0
    human_input = 0
    human_count_input = [0,0,0]
    r = 0 
    p = 0
    s = 0

    print('To play Rock, enter 1')
    print('To play Paper, enter 2')
    print('To play Scissors, enter 3')
    print('To end game, enter 4')

    #run while loop to continue the code

    while (human_input < ROCK or human_input > QUIT):
        human_input = int (input ('Rock, Paper, Scissors? : '))
        
    while (human_input != QUIT):
        

        #call computer to play play 
        computer_result = computer_plays(r, p, s) 
        #make sure to pass in count that users
        
        winner, r, p, s = determine_winner(human_input, computer_result, r, p, s)
        
        #increment # of games
        num_games += 1
            
        
        
        if (winner == HUMAN):
            #if human wins, increment number 
            num_human_wins += 1
            print("You win! Congrats!")
        elif (winner == COMPUTER):
            print("The computer won! Better luck next time!")
        else:
            print("It's a tie!")
            
        human_input = 0
        
        while (human_input < ROCK or human_input > QUIT):
            human_input = int (input ('Rock, Paper, Scissors? : '))
    
    #the user pressed quit
    #print the number of games the user won and
    print
    print("Number of games played: " + str(num_games))
    print("Number of human wins: " + str(num_human_wins))


def determine_winner(human_input, computer_result, r, p, s):
    
        winner = COMPUTER
        
    #put the input into a list
        if(human_input == ROCK):
            #adds count to list 
            r += 1 
            
            if computer_result == SCISSORS:
             #changes winner
                winner = HUMAN 
                
        elif(human_input == PAPER):
            #adds count to list 
            p += 1 
            
            if computer_result == ROCK:
                #changes winner
                winner = HUMAN
                
        
        else:
            #adds count to list 
            s += 1 
            
            if computer_result == PAPER:
                #changes winner
                winner = HUMAN
                
                
        #check to see if tie
        if human_input == computer_result:
            winner = TIE
            
          
        #turn what computer played into string
        if (computer_result == ROCK):
            str_computer_played = 'Rock'
        elif (computer_result == PAPER):
            str_computer_played = 'Paper'
        else:
            str_computer_played = 'Scissors'
            
            
        #print what each player played
        print
        print("Computer Played: " + str_computer_played)
        
        return winner, r, p, s
    
def computer_plays(r, p, s):
    
    #want to understand the algoritm a bit more
    if (r==p and r==s):
        computer_play = random.choice([1,2,3])
    elif (r>=p and r>=s):
        computer_play = PAPER
    elif (p>=r and p>=s):
        computer_play = SCISSORS
    elif (s>=r and s>=p): 
        computer_play = ROCK
   
                
    #return value to play
    return computer_play



main()
```

```python
INPUT = 'movies.csv'


def main():
    
    #open the file
    movies = open(INPUT, 'r')
    
    #call function that makes the dictionary from csv file
    info = make_dict(movies)
    
    #get user input and call function to check
    first_movie = raw_input("First Movie: ") 
    first_movie = check_input(first_movie, info)
    
    second_movie = raw_input("Second Movie: ") 
    second_movie = check_input(second_movie, info)
    
    #get list of actors from each movie
    first_movie_actors = set(info[first_movie])
    second_movie_actors = set(info[second_movie])
    
    #find actors in both 
    intersection = first_movie_actors.intersection(second_movie_actors)
    #find actors in 1 but not 2
    one_not_two = first_movie_actors.difference(second_movie_actors)
    #find actors in 2 but not 1
    two_not_one = second_movie_actors.difference(first_movie_actors)

    print
    print("Actors in both movies: ")
    if len(intersection) == 0:
        print ""
    else:
        #run for loop to print each actor
        for actor in intersection:
            print actor
    
    print 
    print "Actors in Movie 1 NOT Movie 2: "
    for actor in one_not_two:
        print actor
    print 
    print "Actors in Movie 2 NOT Movie 1: "
    for actor in two_not_one:
        print actor
     
    #close the file 
    movies.close()
    
    
def make_dict(movies):
    '''This function takes in the file we opened, reads through each line in the 
    file and puts the movie name as the key and actors as the values. It then 
    returns the dictionary back to the main function'''
    #create an empty dictionary
    info = {}
    
    #read each line and do something
    for line in movies:
        line.strip()
        line.strip("\r\n")
        info_list = line.split(",")
        
        #run a for loop through all the movies provided
        for i in range(1,len(info_list)):
            #check to see if movie in dict
            if info_list[i] in info:
                #if movie already in dict, add actor to list
                info[info_list[i]].append(info_list[0])
                
                #info[info_list[i]] = actor_list.append(info_list[0])
    
                #info[info_list[i]] = info[info_list[i]].append(info_list[0])
            #if movie not in dict, create new listing
            else:
                actor_list = [info_list[0]]
                info[info_list[i]] = actor_list
                
    return info


def check_input(movie_name,info):
    '''takes the user input and the dictionary of movies/actors
    and makes sure that the movie entered is in the dictionary. This 
    function then returns the movie name to the main function.'''
    
    while movie_name not in info:
        movie_name = raw_input("Re-enter Movie Name: ")
        
    return movie_name

main()       
```
